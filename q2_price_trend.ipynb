### Q2.1
House prices are known to change over time due to various factors.

The "Q2 data" worksheet is a table of transaction records for five different highrise buildings.

**Valuation price** refers to the price at which the house is sold, while **Built up price psf** is the Valuation price divided by the area of the house in sq ft (in the Size built up column).

**Quarter** refers to the date at which the property is sold (e.g. 20192 means Quarter 2 of 2019).

The chart above shows the average price psf for these highrise buildings over time.

Choose three buildings and compare amongst them based on the price trend chart and the raw data available in the "Q2 data" worksheet.


Choice: Astra Suites, Lily Point and Rainbow Apartment

```python
price_trend
```

```python
price_trend[price_trend['NAME']=='ASTRA SUITES']
```

```python
price_trend['Valuation price']
```

```python
price_trend.groupby(['NAME','QUARTER'])['Built up price psf'].mean().unstack()
```

```python
plotdata_mean = price_trend.groupby(['NAME','QUARTER'])['Built up price psf'].mean()
```

```python
price_trend.groupby(['NAME','QUARTER'])['Built up price psf'].mean()['ASTRA SUITES']
```

```python
fig, ax = plt.subplots(figsize=(8,6))

plotdata_mean['ASTRA SUITES'].plot(label='ASTRA SUITES')
plotdata_mean['LILY POINT'].plot(label='LILY POINT')
plotdata_mean['RAINBOW APARTMENT'].plot(label='RAINBOW APARTMENT')

plt.title('Average Price psf over Time')
plt.ylabel('Average price psf (RM)')
ax.set_xlim(left=pd.Period(2015), right=pd.Period(2019))
ax.grid(axis = "x")
ax.grid(which = "minor")
# ax.minorticks_on()
plt.legend()

plt.savefig("q2-1", facecolor='w', bbox_inches="tight",
            pad_inches=0.3, transparent=True)

plt.show()
```

### Q2.2						
Plot another price trend that shows the median price psf across time for these three buildings.						
You may use any software/method of your choice.						
What is the difference between the median and average value of a set of data?						


```python
plotdata_median = price_trend.groupby(['NAME','QUARTER'])['Built up price psf'].median()
```

```python
fig, ax = plt.subplots(figsize=(8,6))

plotdata_median['ASTRA SUITES'].plot(label='ASTRA SUITES')
plotdata_median['LILY POINT'].plot(label='LILY POINT')
plotdata_median['RAINBOW APARTMENT'].plot(label='RAINBOW APARTMENT')

plt.title('Median Price psf over Time')
plt.ylabel('Median price psf (RM)')
ax.set_xlim(left=pd.Period(2015), right=pd.Period(2019))
ax.grid(axis = "x")
ax.grid(which = "minor")
plt.legend()

plt.savefig("q2-2-median", facecolor='w', bbox_inches="tight",
            pad_inches=0.3, transparent=True)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,6))

plotdata_mean['ASTRA SUITES'].plot(label='ASTRA SUITES - mean', color='#0CFD21')
plotdata_median['ASTRA SUITES'].plot(label='ASTRA SUITES - median', color='#0A58FF')

plotdata_mean['LILY POINT'].plot(label='LILY POINT - mean', color='#FD0C0C')
plotdata_median['LILY POINT'].plot(label='LILY POINT - median', color='#0CF9FD')

plotdata_mean['RAINBOW APARTMENT'].plot(label='RAINBOW APARTMENT - mean', color='#FD0CF6')
plotdata_median['RAINBOW APARTMENT'].plot(label='RAINBOW APARTMENT - median', color='#FBA000')

plt.title('Difference of Mean and Median Price psf over Time')
plt.ylabel('Median price psf (RM)')
ax.set_xlim(left=pd.Period(2015), right=pd.Period(2019))
ax.grid(axis = "x")
ax.grid(which = "minor")
plt.legend()

plt.savefig("q2-2-diff", facecolor='w', bbox_inches="tight",
            pad_inches=0.3, transparent=True)

plt.show()
```

### Q2.3							
Plot another price trend that shows the median valuation price across time for these three buildings.							
You may use any software/method of your choice.							
Discuss and compare the three buildings using this chart.							


```python
valuation_mean = price_trend.groupby(['NAME','QUARTER'])['Valuation price'].mean()
fig, ax = plt.subplots(figsize=(8,6))

valuation_mean['ASTRA SUITES'].plot(label='ASTRA SUITES')
valuation_mean['LILY POINT'].plot(label='LILY POINT')
valuation_mean['RAINBOW APARTMENT'].plot(label='RAINBOW APARTMENT')

plt.title('Average Price over Time')
plt.ylabel('Mean Valuation price (RM)')
ax.set_xlim(left=pd.Period(2015), right=pd.Period(2019))
ax.grid(axis = "x")
ax.grid(which = "minor")
plt.legend()

plt.savefig("q2-3-mean", facecolor='w', bbox_inches="tight",
            pad_inches=0.3, transparent=True)
plt.show()
```

```python
valuation_median = price_trend.groupby(['NAME','QUARTER'])['Valuation price'].median()
```
